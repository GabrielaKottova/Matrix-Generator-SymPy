# generator


```
import sympy as sp
import random
from sympy import Matrix, symbols

# Toto nastavení zajistí, že se matice a rovnice budou zobrazovat jako v učebnici
sp.init_printing()
```

```
def generujSoustavy(k, n):
    seznam_matic = [] # Připravíme si prázdný seznam na hotové příklady

    for j in range(k): # Tento cyklus vyrobí přesně k příkladů
        # 1. Základ - jednotková matice
        matice_A = sp.eye(n)

        # 2. Míchání (dáme rovnou 10 míchání, ať to vypadá složitě)
        pocet_michani = 12
        for i in range(pocet_michani):
            radek_kam, radek_odkud = random.sample(range(n), 2)
            nasobek = random.randint(-3, 3)
            if nasobek == 0: nasobek = 1
            matice_A[radek_kam, :] = matice_A[radek_kam, :] + nasobek * matice_A[radek_odkud, :]

        # 3. Vymyšlení výsledků (to, co pak vyjde za x, y...)

        reseni = Matrix([random.randint(-5, 5) for _ in range(n)])
        print(f"výsledek pro {j + 1}. příklad je: {list(reseni)}")


        # 4. Dopočítání pravé strany (to, co je za rovnítkem)
        vektor_B = matice_A * reseni

        # 5. SLEPENÍ: Spojíme matici A a vektor B do jedné velké tabulky
        rozsirena_matice = matice_A.row_join(vektor_B)

        # Hotovou tabulku vložíme do našeho seznamu
        seznam_matic.append(rozsirena_matice)

    # Až automat vyrobí k příkladů, celou krabici nám vrátí
    return seznam_matic

# --- Konec funkce, teď si ji jdeme vyzkoušet ---

# Nastavíme si k a n do proměnných
pocet_k = 3 # Zkusíme schválně 3 příklady!
velikost_n = 3

moje_priklady = generujSoustavy(k=pocet_k, n=velikost_n)

# TADY JE TA ZMĚNA: Místo ručního vypisování použijeme cyklus!
for i in range(pocet_k):
    print(f"\nTady je {i + 1}. vygenerovaný příklad (rozšířená matice):")
    display(moje_priklady[i])
```

```
print(f"Tady jsou krásně vypsané ÚPLNĚ VŠECHNY tvé soustavy rovnic:")

# Cyklus, který projde všechny vygenerované tabulky v tvém seznamu
for cislo_prikladu, matice_ukazka in enumerate(moje_priklady):

    # Samo si to zjistí velikost aktuální tabulky
    pocet_n = matice_ukazka.rows

    # Vyrobí si to správný počet písmenek
    x = sp.symbols(f'x1:{pocet_n + 1}')

    # Vypíšeme si hezký nadpis pro každý příklad
    print(f"\n \n {cislo_prikladu + 1}. SOUSTAVA ROVNIC: \n")

    # Poskládáme a vykreslíme rovnice
    for i in range(pocet_n):
        leva_strana = sum(matice_ukazka[i, j] * x[j] for j in range(pocet_n))
        prava_strana = matice_ukazka[i, pocet_n]
        display(sp.Eq(leva_strana, prava_strana))
```
